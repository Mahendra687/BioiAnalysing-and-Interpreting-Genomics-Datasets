### Week2: WfMS and Analysis

#### Installing Nextflow:
- Check the JAVA version command: $ `java -version`

- Download and install Nextflow command: $ `curl -fsSL get.nextflow.io | bash`

- Now, you need to ensure that the directory is exists. before running the `mv` command if the directory does not exist then you can create it as well for example, I have created a directory for this course by the following command: $ `mkdir ~/MOOC`

- Now, move the nextflow executable command: $ `mv nextflow ~/MOOC/`

- After moving the nextflow on to MOOC directory we have to navigate the directory using the command: $ `cd MOOC/`

- To check it’s working, run the following command: ~/MOOC$ `./nextflow help`

#### Through conda:
- command: $ `conda create --name nextflow nextflow`

- command: $ `conda activate nextflow`

- command: $ `nextflow help`

#### Install Singularity
- command: $ `sudo apt install -y runc cryptsetup-bin`

- command: $ `wget -O singularity.deb https://github.com/sylabs/singularity/releases/download/v3.11.4/singularity-ce_3.11.4-jammy_amd64.deb`

- command: $ `sudo dpkg -i singularity.deb`

- command: $ `rm singularity.deb`

#### In case error Fix-broken install:

- Command: $ `sudo apt --fix-broken install`

- Command: $ `sudo apt install -y uidmap`

- Command: $ `sudo apt install -y singularity-ce`


#### The nf-core viralrecon pipeline

- **Creating a samplesheet for viralrecon:**
- command: $ `conda activate MOOC` to activate coda
- create a text file called samples.txt including the following ENA accessions: ERR5556343, SRR13500958, ERR5743893, ERR5181310 and ERR5405022

- use a for loop to run fastq-dump on each of the accessions in samples.txt: using the following command: $ `for i in $(cat samples.txt);do fastq-dump --split-files $i;done`

- compress the fastq files we just downloaded with gzip using the command: $ `gzip *.fastq`
- create a directory called data and move the fastq files there:

> - command: $ `mkdir data`

> - command: $ `mv *.fastq.gz data`

#### Running viralrecon

command: $ `conda activate nextflow`

Command: $ `nextflow run nf-core/viralrecon -profile singularity \`

`--max_memory '12.GB' --max_cpus 4 \`

`--input samplesheet.csv \`

`--outdir results/viralrecon \`

`--protocol amplicon \`

`--genome 'MN908947.3' \`

`--primer_set artic \`

`--primer_set_version 3 \`

`--skip_kraken2 \`

`--skip_assembly \`

`--skip_pangolin \`

`--skip_nextclade \`

`--platform illumina`
